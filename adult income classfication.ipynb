# preprocessing

```python
x=df.drop('income',axis=1)
y=df['income']
```

```python
y = y.map({'<=50K': 0, '>50K': 1})
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 0)
```

```python
num_scaled=x_train.select_dtypes(include='number').columns
```

```python
from sklearn.preprocessing import RobustScaler
rc=RobustScaler()
x_train[num_scaled]=rc.fit_transform(x_train[num_scaled])
x_test[num_scaled]=rc.transform(x_test[num_scaled])
```

```python
x_train.head()
```

```python
nominal_cols = x_train.select_dtypes(include= 'object').columns
```

```python
for col in nominal_cols:
    print(col)
    print(df[col].nunique())
```

```python
nominal_cols_ohe=x_train.select_dtypes(include= 'object').drop(['occupation','native-country'], axis = 1).columns
```

```python
nominal_cols_bin=['occupation','native-country']
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output= False, drop= 'first')

ohe_x_train = ohe.fit_transform(x_train[nominal_cols_ohe])
ohe_x_test = ohe.transform(x_test[nominal_cols_ohe])
```

```python
ohe_x_train_df = pd.DataFrame(ohe_x_train, columns= ohe.get_feature_names_out())
ohe_x_test_df = pd.DataFrame(ohe_x_test, columns= ohe.get_feature_names_out())
ohe_x_train_df
```

```python
import category_encoders as ce
binary_encoder = ce.BinaryEncoder(cols=nominal_cols_bin)
bin_X_train = binary_encoder.fit_transform(x_train[nominal_cols_bin])
bin_X_test = binary_encoder.fit_transform(x_test[nominal_cols_bin])
```

```python
bin_X_train
```

```python
x_train.reset_index(inplace= True, drop= True)
x_test.reset_index(inplace= True, drop= True)

y_train.reset_index(inplace= True, drop= True)
y_test.reset_index(inplace= True, drop= True)
```

```python
bin_X_train.reset_index(inplace= True, drop= True)
bin_X_test.reset_index(inplace= True, drop= True)
ohe_x_train_df.reset_index(inplace= True, drop= True)
ohe_x_test_df.reset_index(inplace= True, drop= True)
```

```python
x_train = pd.concat([x_train,ohe_x_train_df,bin_X_train ], axis= 1).drop(nominal_cols, axis= 1)

x_test = pd.concat([x_test, ohe_x_test_df,bin_X_test], axis= 1).drop(nominal_cols, axis= 1)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(penalty= None, random_state= 42)

lr.fit(x_train, y_train)

print('Training Accuracy :', lr.score(x_train, y_train) * 100)
print('Test Accuracy :', lr.score(x_test, y_test) * 100)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(penalty= 'l2', C= .5, random_state= 42)

lr.fit(x_train, y_train)

print('Training Accuracy :', lr.score(x_train, y_train) * 100)
print('Test Accuracy :', lr.score(x_test, y_test) * 100)
```
